```python
import pandas as pd
import requests
import json
```

```python
df=pd.read_csv("resources//API_EN.ATM.CO2E.KT_DS2_en_csv_v2_3470002.csv") 
```

```python
df.head()
```

```python
df = df.drop(df.columns[[2,3]], axis = 1)
```

```python
df_rotated = df.melt(id_vars=["Country Name", "Country Code"], 
        var_name="Year", 
        value_name="CO2 Emissions KT")
```

```python
df_rotated.head()
```

```python
from datetime import datetime
timestampStr = datetime.now().date().strftime("%d-%b-%Y")

parsedDf = json.loads((df_rotated).to_json(orient="records"))

with open('emissions_data_by_country_1960-2020-'+timestampStr+'.json', 'w') as outfile:
     json.dump(parsedDf , outfile, sort_keys = True, indent = 4,
               ensure_ascii = False)
```
